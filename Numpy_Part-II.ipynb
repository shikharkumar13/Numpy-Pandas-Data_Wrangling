## Importing and reading data from files

We don't usually create the arrays that we work with. We need to be able to import data present in another file to Numpy.

Numpy provides useful functions to load data from an external file and save it as well.

The main import methods are:

1. numpy.loadtxt()
2. numpy.genfromtext()

Use `np.loadtxt` when there is no missing value in the data.

```python
import numpy as np
```

```python
data = np.loadtxt('data.txt', delimiter="\t")
data
```

```python
type(data)
```

**But this function throws error when reading files with missing values**

```python
data = np.loadtxt('data_miss.txt', delimiter="\t")
data
```

In such situation, use `np.genfromtxt()`. It fills in missing data with `nan`.

```python
data = np.genfromtxt('data_miss.txt', delimiter="\t")
data
```

# Business Problem Statements using Numpy

### The Ride-Sharing App (Urban Mobility Analytics)

You are a Data Scientist at a major ride-sharing company (like Uber or Lyft). The city operations team has given you a batch of ride data from a busy Friday night. They need deep insights into driver efficiency, surge pricing dynamics, and rider satisfaction to optimize the matching algorithm.

1.1 **Data Ingestion**: Load the data into a NumPy array.

```python
data = np.genfromtxt('ride_data.csv', delimiter=',', skip_header=1)

ride_ids = data[:, 0]
driver_ids = data[:, 1]
distances = data[:, 2]
durations = data[:, 3]
fares = data[:, 4]
surges = data[:, 5]
d_ratings = data[:, 6]
r_ratings = data[:, 7]
```

**These are the columns in the data**

ride_id, driver_id, distance_miles, duration_minutes, fare_usd, surge_multiplier, driver_rating, rider_rating

- `ride_id` : The unique id given to every ride.
- `driver_id` : The unique driver id of each driver.
- `distance_miles` : Distance covered from origin to destination (in miles).
- `duration_minutes` : Duration of each ride in minutes.
- `fare_usd` : Total fare of the ride is US Dollars.
- `surge_multiplier` : surge pricing will activate for a ride during a certain event and the user will have to pay a higher amount. For example, if the surge multiplier is 1.2, then the `total fare will be normal ride fare * 1.2`.
- `driver_rating` : Rating given to the driver by the user.
- `rider_rating` : Rating given to rider by the driver for a trip

```python
data
```

2.1 **Gross Merchandise Value (GMV)**: Calculate the total fares collected.

2.2 **Platform Revenue**: The company takes a 25% commission. Calculate total platform revenue.

```python
gmv = np.sum(fares)
platform_revenue = gmv * 0.25
print(f"Total GMV: ${gmv:.2f} | Platform Cut: ${platform_revenue:.2f}")
```

This means, for all rides in the data, company collected total fare of `$403.50`. And since they keep only 25% commission from it, their total revenue was `$100.88`.

The GMV is the sum of every single dollar paid by every single rider before the company and the driver gets their cut.

Imagine you build an app that connects local bakers with customers. A customer buys a `$100` wedding cake. You take a `$10` fee for connecting them, and the baker keeps $90.

- Your Revenue is $10.

- But the GMV of that transaction is $100.

**How and why it helps a company:**
In Big Tech (Uber, Airbnb, Amazon, eBay), GMV is the ultimate vanity metric for scale and market share. Investors look at GMV to see how much economic activity the platform is generating.

- If a ride-sharing app has a GMV of $1 Billion, it means they control a massive chunk of city transportation.

- `It also dictates their potential ceiling`: you cannot increase your revenue without either taking a bigger percentage cut (which angers drivers) or increasing the overall GMV.

3.1 **Average Speed**: Calculate the average speed (mph) for every ride to check for traffic anomalies.

3.2 **Speeding Flag**: Identify ride_ids where the average speed exceeded 65 mph.

```python
speeds_mph = distances / (durations / 60)
speeding_rides = ride_ids[speeds_mph > 65]
print("\nSpeeds (mph):\n", np.round(speeds_mph, 1))
print("Speeding Ride IDs:", speeding_rides.astype(int))
```

4.1 **Surge Impact Calculation**: How much extra money was generated purely because of surge multipliers?

```python
base_fares = fares / surges
surge_extra_revenue = np.sum(fares - base_fares)
print(f"\nExtra Revenue from Surge: ${surge_extra_revenue:.2f}")
```

Surge pricing is a dynamic pricing model. A Surge Multiplier is the exact number the base fare is multiplied by (e.g., 1.5x, 2x) during periods of high demand.

**Why it is implemented:**
Surge pricing exists to solve a simple supply-and-demand problem. When demand is much more than supply, the system breaks. Surge pricing does two things simultaneously:

- `Incentivizes or Encourages Supply of Drivers`: It alerts drivers ("There is a 2.0x surge in that area") prompting more drivers to log on or drive toward that the area of high demand.

- `Controls Demand`: It discourages riders who don't urgently need a ride. Someone might decide to wait 20 minutes or take the bus, freeing up the available cars for people who want it more urgently and are willing to pay the premium (like someone rushing to the airport).

7. **Short-Trip Ratio**: What percentage of rides were under 3 miles?

```python
short_trips_pct = np.mean(distances < 3.0) * 100
print(f"\nShort Trips (< 3 miles): {short_trips_pct:.1f}%")
```

8. **Rating Discrepancy**: Calculate the absolute difference between the driver's rating and the rider's rating for each trip.

```python
rating_diff = np.abs(d_ratings - r_ratings)
print("\nAbsolute Rating Discrepancies:\n", np.round(rating_diff, 1))
```

9. **High-Value Threshold**: Find the 90th percentile fare amount.

The `High-Value Threshold` is the dollar amount that separates the top 10% most expensive rides from the bottom 90% of normal rides.

If the 90th percentile is `$42.00`, it means `90%` of all rides cost less than `$42`, and only the top 10% cost more.

```python
p90_fare = np.percentile(fares, 90)
print(f"\n90th Percentile Fare Threshold: ${p90_fare:.2f}")
```

A ride might cross this threshold because it’s a very long distance (an airport drop-off), it happened during a massive 3.0x surge, or the rider booked a luxury SUV (Uber Black).

**Why Data Scientists calculate it:**

You calculate this threshold to segment your data and treat those "premium" trips differently.

- **For Operations**: You want to make sure your absolute best, 5-star drivers are the ones being dispatched to these high-value riders to ensure a flawless experience.

- **For Marketing**: If you know the threshold is `$42`, you can run a campaign: "Spend over `$45` on a ride and get a free coffee voucher."

- **For Anomaly Detection**: If a ride is way, way past this threshold (e.g., $400), a Data Scientist might flag it for review to ensure the GPS didn't glitch or the driver didn't intentionally take the rider in circles to rack up the bill.


10. **Efficiency Metric**: Calculate the dollars earned per minute for each ride.

```python
dollars_per_minute = fares / durations
print("\nDollars earned per minute:\n", np.round(dollars_per_minute, 2))
```

11. **Perfect Match Filter**: Find ride_ids where both driver and rider gave a 5.0 rating.

```python
perfect_mask = (d_ratings == 5.0) & (r_ratings == 5.0)
print("\nPerfect Match Ride IDs:", ride_ids[perfect_mask].astype(int))
```

12. **Driver Penalty**: Create an array of driver payouts (75% of fare). If the driver rating is below 4.0, deduct a flat $3 penalty using np.where.

```python
standard_payout = fares * 0.75
adjusted_payout = np.where(d_ratings < 4.0, standard_payout - 3.00, standard_payout)
print("\nAdjusted Payouts (First 5):\n", np.round(adjusted_payout[:5], 2))
```

13. **Cumulative Fares**: Calculate the running total (cumulative sum) of fares as the night progressed.

```python
cum_fares = np.cumsum(fares)
print("\nRunning Total of Fares:\n", np.round(cum_fares, 2))
```

```python
unique_drivers, counts = np.unique(driver_ids, return_counts=True)
busiest_driver = unique_drivers[np.argmax(counts)]
print(f"\nBusiest Driver ID: {int(busiest_driver)}")
```

15. **Duration Sorting**: Get the list of ride_ids sorted from shortest duration to longest duration.

```python
sorted_ride_ids = ride_ids[np.argsort(durations)]
print("\nRide IDs sorted by duration (fastest to longest):", sorted_ride_ids.astype(int))
```

On the surface it seems like such a basic thing to keep track of. But business wise it makes a lot of sense and help a company. How?

When you sort rides by duration, you are immediately identifying two highly problematic categories for the business: `the Micro-Trips (the shortest rides)` and the `Out-of-Position Trips (the longest rides)`.

**1. Analyzing "Micro-Trips" (The Shortest Durations)**
When you look at the top of your sorted list (rides taking 2 to 5 minutes), you are looking at trips that are often highly inefficient for cars.

- **The Driver Frustration Problem**: Imagine a driver spends 8 minutes fighting through city traffic just to pick up a passenger who only wants to go few kms (a 2-minute ride). The driver earns very little and feels like they wasted their time.

- **The Business Solution**: By identifying the volume of these micro-trips, the platform can make strategic decisions. They might implement a "Minimum Fare" to ensure drivers earn a fair amount for their time. Alternatively, if the data shows thousands of these 3-minute rides in a specific area, the company might deploy fleets of electric scooters or e-bikes there, prompting the user in the app: "A car is `$8` and takes 5 mins. A scooter is `$2` and takes 4 mins."

**2. Managing "Deadheading" (The Longest Durations)**
When you look at the very bottom of your sorted list (rides taking 45+ minutes), you are looking at rare outlier rides. They generate a lot of revenue, but they create a massive logistical headache.

- **The "Deadhead" Problem**: A 60-minute ride from the busy city center is great money. But once the driver drops the passenger off, they are stuck in a location where there is no demand. They have to drive 60 minutes back to the city empty. In the logistics world, driving empty is called "deadheading," and it means the driver is burning petrol and losing money.

- **The Business Solution**: By analyzing these long-duration rides, platforms developed features like the "Destination Filter." If a driver gets taken 60 minutes out of the city, the algorithm will specifically look for a passenger outside the city who wants to travel back towards the city, offering it to that specific driver so they get paid for their return trip. Platforms also sometimes add a "Long Trip Premium" fee to the rider to compensate the driver for the empty return journey.
